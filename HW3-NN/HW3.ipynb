In this assignment, NN and SVM are implemented to classify iris data to demonstrate the abilities of non-linearity models in pracitice. NN with Sigmoid and ReLu acitivation functions are utilized to compare the performance. In contrast, SVM with RBF kernel is a common choice for many newbies in this field. However, to illlustrate the performance of different kernels, we gonna apply many different types of kernels such as exponential kernels, RBF. 

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(x):
    x = 1/(1+np.exp(-x))
    return x

def sigmoid_grad(x):
    return (x)*(1-x)

def ReLU(x):
    return np.maximum(0,x)
```

```python
# Neural Networks - using 5-fold cross validation to find hyperparameters
# X_train : input - 120*4 matrix data
# Y_train : one-hot encoding output [1 0 0] - denote that the output will be Sentosa flowers
# X_test  : 30*4 matrix data
# Y_test  : one-hot encoding output

def three_layer_net(NON_LINEARITY, X, Y, model, step_size, reg = 0.0):
    
    h = model['h']
    h2 = model['h2']
    W1 = model['W1']
    W2 = model['W2']
    W3 = model['W3']
    b1 = model['b1']
    b2 = model['b2']
    b3 = model['b3']
    
    num_examples = X.shape[0]
#     print(num_examples)
    plot_array_1 = []
    plot_array_2 = []
    
    for i in range(50000):
        #forward pass
        if NON_LINEARITY == 'ReLU':
            hidden_layer = ReLU(np.dot(X, W1) + b1)
            hidden_layer_2 = ReLU(np.dot(hidden_layer, W2) + b2)
            scores = np.dot(hidden_layer_2, W3) + b3
        elif NON_LINEARITY == 'SIGM':
            hidden_layer = sigmoid(np.dot(X,W1) + b1)
            hidden_layer_2 = sigmoid(np.dot(hidden_layer, W2) + b2)
            scores = np.dot(hidden_layer_2, W3) + b3

        #softmax function to classify 
        exp_scores = np.exp(scores)
        probs = exp_scores/ np.sum(exp_scores, axis = 1, keepdims = True)

        #print(num_examples)
        correct_log_probs = -np.log(probs[range(num_examples), Y])
        data_loss = np.sum(correct_log_probs)/num_examples
        reg_loss = 0.5*reg*np.sum(W1*W1) + 0.5*reg*np.sum(W2*W2) + 0.5*reg*np.sum(W3*W3)
        loss = data_loss + reg_loss
        if i % 1000 == 0:
            print("iteration %d: loss %f" % (i, loss))
        #if i == 1000:
        #    print("W1: ", W1)
        #    print("W2: ", W2)
        #    print("W3: ", W3)
            #print("scores", scores)
        #    print("exp_scores",exp_scores)
            #print("correct", correct_log_probs)
        #    print("reg_loss", reg_loss)
        #    print("data_loss", data_loss)
            
        # compute the gradient on scores
        dscores = probs
        dscores[range(num_examples),Y] -= 1
        dscores /= num_examples
        print(dscores.shape)
        
        dW3 = np.dot(hidden_layer_2.T, dscores) + reg*W3
        db3 = np.sum(dscores, axis=0, keepdims=True)
        
        if NON_LINEARITY == 'ReLU':
            #backprop ReLU nonlinearity here
            dhidden2 = np.dot(dscores, W3.T)
            dhidden2[hidden_layer_2 <= 0] = 0
            dW2 =  np.dot( hidden_layer.T, dhidden2) + reg*W2
            #plot_array_2.append(np.sum(np.abs(dW2))/np.sum(np.abs(dW2.shape)))
            db2 = np.sum(dhidden2, axis=0)
            dhidden = np.dot(dhidden2, W2.T)
            dhidden[hidden_layer <= 0] = 0
            
            
        elif NON_LINEARITY == 'SIGM':
            #backprop sigmoid nonlinearity here
            dhidden2 = dscores.dot(W3.T)*sigmoid_grad(hidden_layer_2)
            dW2 = (hidden_layer.T).dot(dhidden2) + reg*W2
            plot_array_2.append(np.sum(np.abs(dW2))/np.sum(np.abs(dW2.shape)))
            db2 = np.sum(dhidden2, axis=0)
            dhidden = dhidden2.dot(W2.T)*sigmoid_grad(hidden_layer)

            
            
               
        dW1 = np.dot(X.T, dhidden) + reg*W1
        db1 = np.sum(dhidden, axis=0)
        
        #option to return loss, grads -- uncomment next comment
        grads={}
        grads['W1']=dW1
        grads['W2']=dW2
        grads['W3']=dW3
        grads['b1']=db1
        grads['b2']=db2
        grads['b3']=db3
        #return loss, grads
        
        #Update
        W1 += -step_size*dW1
        b1 += -step_size*db1
        W2 += -step_size*dW2
        b2 += -step_size*db2
        W3 += -step_size*dW3
        b3 += -step_size*db3
    
    #After training
    #Evaluate training accuracy
    if NON_LINEARITY == 'ReLU':
        hidden_layer = ReLU(np.dot(X, W1) + b1)
        hidden_layer_2 = ReLU(np.dot(hidden_layer, W2) + b2)
    elif NON_LINEARITY == 'SIGM':
        hidden_layer = sigmoid(np.dot(X, W1) + b1)
        hidden_layer_2 = sigmoid(np.dot(hidden_layer, W2) + b2)
    
    scores = np.dot(hidden_layer_2, W3) + b3
    predicted_class = np.argmax(scores, axis=1)
    
    print('training accuracy: %.2f' % (np.mean(predicted_class == Y)))
    
    return W3, W2, W1, b3, b2, b1
        
```

```python
#function to train a three layer neural net with either RELU or sigmoid nonlinearity via vanilla grad descent

def three_layer_net_new(NONLINEARITY,X,y, model, step_size, reg):
    #parameter initialization
    
    h= model['h']
    h2= model['h2']
    W1= model['W1']
    W2= model['W2']
    W3= model['W3']
    b1= model['b1']
    b2= model['b2']
    b3= model['b3']
    
    
    # some hyperparameters


    # gradient descent loop
    num_examples = X.shape[0]
    plot_array_1=[]
    plot_array_2=[]
    for i in range(50000):

        #FOWARD PROP

        if NONLINEARITY== 'ReLU':
            hidden_layer = ReLU(np.dot(X, W1) + b1)
            hidden_layer2 = ReLU(np.dot(hidden_layer, W2) + b2)
            scores = np.dot(hidden_layer2, W3) + b3
            
        elif NONLINEARITY == 'SIGM':
            hidden_layer = sigmoid(np.dot(X, W1) + b1)
            hidden_layer2 = sigmoid(np.dot(hidden_layer, W2) + b2)
            scores = np.dot(hidden_layer2, W3) + b3

        exp_scores = np.exp(scores)
        probs = exp_scores / np.sum(exp_scores, axis=1, keepdims=True) # [N x K]
        
        #print(num_examples)
        # compute the loss: average cross-entropy loss and regularization
        corect_logprobs = -np.log(probs[range(num_examples),y])
        data_loss = np.sum(corect_logprobs)/num_examples
        reg_loss = 0.5*reg*np.sum(W1*W1) + 0.5*reg*np.sum(W2*W2)+ 0.5*reg*np.sum(W3*W3)
        loss = data_loss + reg_loss
        if i % 1000 == 0:
            print(corect_logprobs.shape)
            print("iteration %d: loss %f" % (i, loss))


        # compute the gradient on scores
        dscores = probs
        dscores[range(num_examples),y] -= 1
        dscores /= num_examples
        print(dscores.shape)

 
        # BACKPROP HERE
        dW3 = (hidden_layer2.T).dot(dscores)
        db3 = np.sum(dscores, axis=0, keepdims=True)


        if NONLINEARITY == 'ReLU':

            #backprop ReLU nonlinearity here
            dhidden2 = np.dot(dscores, W3.T)
            dhidden2[hidden_layer2 <= 0] = 0
            dW2 =  np.dot( hidden_layer.T, dhidden2)
            plot_array_2.append(np.sum(np.abs(dW2))/np.sum(np.abs(dW2.shape)))
            db2 = np.sum(dhidden2, axis=0)
            dhidden = np.dot(dhidden2, W2.T)
            dhidden[hidden_layer <= 0] = 0
            
        elif NONLINEARITY == 'SIGM':

            #backprop sigmoid nonlinearity here
            dhidden2 = dscores.dot(W3.T)*sigmoid_grad(hidden_layer2)
            dW2 = (hidden_layer.T).dot(dhidden2)
            plot_array_2.append(np.sum(np.abs(dW2))/np.sum(np.abs(dW2.shape)))
            db2 = np.sum(dhidden2, axis=0)
            dhidden = dhidden2.dot(W2.T)*sigmoid_grad(hidden_layer)
                
        dW1 =  np.dot(X.T, dhidden)
        plot_array_1.append(np.sum(np.abs(dW1))/np.sum(np.abs(dW1.shape)))
        db1 = np.sum(dhidden, axis=0)

        # add regularization
        dW3+= reg * W3
        dW2 += reg * W2
        dW1 += reg * W1
        
        #option to return loss, grads -- uncomment next comment
        grads={}
        grads['W1']=dW1
        grads['W2']=dW2
        grads['W3']=dW3
        grads['b1']=db1
        grads['b2']=db2
        grads['b3']=db3
        #return loss, grads
        
        
        # update
        W1 += -step_size * dW1
        b1 += -step_size * db1
        W2 += -step_size * dW2
        b2 += -step_size * db2
        W3 += -step_size * dW3
        b3 += -step_size * db3
    
    #After training
    # evaluate training set accuracy
    if NONLINEARITY == 'ReLU':
        hidden_layer = ReLU(np.dot(X, W1) + b1)
        hidden_layer2 = ReLU(np.dot(hidden_layer, W2) + b2)
    elif NONLINEARITY == 'SIGM':
        hidden_layer = sigmoid(np.dot(X, W1) + b1)
        hidden_layer2 = sigmoid(np.dot(hidden_layer, W2) + b2)
        
    scores = np.dot(hidden_layer2, W3) + b3
    predicted_class = np.argmax(scores, axis=1)
    print('training accuracy: %.2f' % (np.mean(predicted_class == y)))
    #return cost, grads
    return W3, W2, W1, b3, b2, b1
```

```python
#loading iris data via panda
import pandas as pd

dataIris = pd.read_table('iris.dat', delim_whitespace = True, header = None)
dataIris.head()

arrayIrisData = dataIris.as_matrix()

irisSetosa = arrayIrisData[0:50, 0:4]
irisVerisColor = arrayIrisData[50:100, 0:4]
irisVirginica  = arrayIrisData[100:150, 0:4]

train_mats1 = [irisSetosa[10:50,:], irisVerisColor[10:50,:], irisVirginica[10:50,:]]
test_mats1  = [irisSetosa[0:10,:],  irisVerisColor[0:10,:], irisVirginica[0:10,:]]

train_mats2 = [np.concatenate((irisSetosa[0:10, :], irisSetosa[20:50,:]), axis = 0), 
               np.concatenate((irisVerisColor[0:10,:], irisVerisColor[20:50,:]), axis = 0), 
               np.concatenate((irisVirginica[0:10,:], irisVirginica[20:50,:]), axis = 0)]
test_mats2  = [irisSetosa[10:20,:], irisVerisColor[10:20,:], irisVirginica[10:20,:]]

train_mats3 = [np.concatenate((irisSetosa[0:20, :], irisSetosa[30:50,:]), axis = 0), 
               np.concatenate((irisVerisColor[0:20,:], irisVerisColor[30:50,:]), axis = 0), 
               np.concatenate((irisVirginica[0:20,:], irisVirginica[30:50,:]), axis = 0)]
test_mats3  = [irisSetosa[20:30,:], irisVerisColor[20:30,:], irisVirginica[20:30,:]]

train_mats4 = [np.concatenate((irisSetosa[0:30, :], irisSetosa[40:50,:]), axis = 0), 
               np.concatenate((irisVerisColor[0:30,:], irisVerisColor[40:50,:]), axis = 0), 
               np.concatenate((irisVirginica[0:30,:], irisVirginica[40:50,:]), axis = 0)]
test_mats4  = [irisSetosa[30:40,:], irisVerisColor[30:40,:], irisVirginica[30:40,:]]

train_mats5 = [irisSetosa[0:40,:], irisVerisColor[0:40,:], irisVirginica[0:40,:]]
test_mats5  = [irisSetosa[40:50,:], irisVerisColor[40:50,:],irisVirginica[40:50,:]]


#train_mats
train_mats = [train_mats1, train_mats2, train_mats3, train_mats4, train_mats5]
test_mats  = [test_mats1, test_mats2, test_mats3, test_mats4, test_mats5]

#irisSetosa.shape
#train_mats1[0].shape

#Y -target value
Y =  np.zeros(120*1, dtype = 'uint8')
Y[40:80] = 1
Y[80:120]= 2
print(Y.shape)
```

```python
#Initialize neural networks models parameters
D = 4 #input D-dimensional vector
K = 3 #number of output classes
h = 50
h2 = 30

model = {}
model['h'] = h
model['h2'] = h2
model['W1'] = 0.1*np.random.randn(D, h)
model['b1'] = np.zeros((1,h))
model['W2'] = 0.1*np.random.randn(h, h2)
model['b2'] = np.zeros((1, h2))
model['W3'] = 0.1*np.random.randn(h2, K)
model['b3'] = np.zeros((1, K))
```

```python
def predict(W3, W2, W1, b3, b2, b1, X, NON_LINEARITY):
    #data_input : N*D matrix - N number of samples, D-dimesional vector
    #data_output: 
    if NON_LINEARITY == 'ReLU':
        hidden_layer = ReLU(np.dot(X, W1)) + b1
        hidden_layer_2 = ReLU(np.dot(hidden_layer, W2)) + b2
    elif NON_LINEARITY == 'SIGM':
        hidden_layer = sigmoid(np.dot(X, W1)) + b1
        hidden_layer_2 = sigmoid(np.dot(hidden_layer, W2)) + b2
        
    scores = np.dot(hidden_layer_2, W3) + b3
    
    exp_scores = np.exp(scores)
    probs = exp_scores / np.sum(exp_scores, axis=1, keepdims=True)
    
    c = [0, 0, 0]
    
    for j in range(10):
        c[np.argmax(probs[j, :])] += 1
        
    return c
            

#Question 1: using NN to classify data 
def classify_q1(NON_LINEARITY, train_mats, test_mats):
    
    #concatenate all training data to a single numpy array
    X = np.concatenate((train_mats[0], train_mats[1], train_mats[2]), axis = 0)
    print("shape of X : ",X.shape)
    print("shape of Y : ", Y.shape)
    
    (W3, W2, W1, b3, b2, b1) = three_layer_net(NON_LINEARITY, X, Y, model, step_size = 1e-3, reg = 1e-5)
    
    print("Confusion matrix: ")
    
    for i in range(3):
        c = predict(W3, W2, W1, b3, b2, b1, test_mats[i], NON_LINEARITY)
        print(c)
    return

def question1(NON_LINEARITY, train, test):
    for i in range(5):
        classify_q1(NON_LINEARITY, train[i], test[i])
    
```
